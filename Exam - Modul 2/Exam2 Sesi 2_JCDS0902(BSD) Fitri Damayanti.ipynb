```python
import numpy as np
```

```python
outliers = []
def outlier_iqr(data):
    sorted(data)
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 -(1.5 * iqr) 
    upper_bound = q3 +(1.5 * iqr)
    for i in data:
        if i < lower_bound or i > upper_bound:
            outliers.append(i)
    return outliers
outlier_datapoints = outlier_iqr(df['price'])
print('Data outlier pada price: ', outlier_datapoints)
```
